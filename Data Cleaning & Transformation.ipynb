```python
#Nous devons maintenant étudier l'abondance de -1 dans la colonne 'Competitors'. 
#Quel est le pourcentage de la colonne entière et combien y en a-t-il ?

df.loc[df["Competitors"] == "-1","Competitors"].value_counts()
```

```python
#il y a 501 occurrences de -1 dans la colonne, ce qui représente 75 % de la colonne ; 
#il est donc préférable de supprimer la colonne puisqu'elle est principalement composée de zéros.

df.drop(columns="Competitors",inplace=True)
```

```python
#Vérification colonne par colonne pour voir si certains éléments doivent être nettoyé. Les 100 premières lignes devraient être bonnes pour un contrôle.

df["Job Title"][:100]
```

```python
""" Après vérification, la colonne 'Job Title' & ' Job Description' présentent des éléments tel que: (), /,\n,et des nombres aleatoires
qui doivent être supprimer, sur certaines lignes, localisons-les  """

df.loc[df["Job Title"].str.contains('\([^()]+\)','\n.*'), ["Job Title",'Job Description']]
```

```python
df['Job Description'] = df['Job Description'].str.replace('\n.*', '')
```

```python
df['Job Title'] = df['Job Title'].str.replace('\n.*','')
```

```python
""" On remarque également la présence de parenthèses avec la mention Sr à l'intérieur, ou encore des infos qui peuvent être pertinentes"""

# remplaçons d'abord (Sr.) par Sr

df["Job Title"]= df.loc[:,"Job Title"].str.replace("(Sr.)","Sr.")
```

```python
# Supprimons le reste des instances entre parenthèses

df["Job Title"]=df.loc[:,"Job Title"].str.extract('([^()]+)')
```

```python
# Vérification pour voir si tout a été supprimer
df.loc[df["Job Title"].str.contains('\([^()]+\)'), ["Job Title","Job Description"]]
```
